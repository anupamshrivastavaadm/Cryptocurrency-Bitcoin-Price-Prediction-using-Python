#Importing libraries and packages
###Latest collection of Bitcoin prices for the past 365 days

```python
import pandas as pd
import yfinance as yf
import datetime
from datetime import date, timedelta
today = date.today()

d1 = today.strftime("%Y-%m-%d")
end_date = d1
d2 = date.today() - timedelta(days=365)
d2 = d2.strftime("%Y-%m-%d")
start_date = d2

data = yf.download('BTC-USD', 
                      start=start_date, 
                      end=end_date, 
                      progress=False)
data["Date"] = data.index
data = data[["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"]]
data.reset_index(drop=True, inplace=True)
print(data.head())
```

Dataset contains 365 rows and 7 columns.

```python
data.shape
```

 Visualizing the changes in bitcoin prices till today by using a candlestick chart.

```python
import plotly.graph_objects as go
figure = go.Figure(data=[go.Candlestick(x=data["Date"],
                                        open=data["Open"], 
                                        high=data["High"],
                                        low=data["Low"], 
                                        close=data["Close"])])
figure.update_layout(title = "Bitcoin Price Analysis", 
                     xaxis_rangeslider_visible=True)
figure.show()
```

Correlation of all the columns in the data

```python
correlation = data.corr()
print(correlation["Close"].sort_values(ascending=False))
```

AutoTS Library in Python is the best libraries for time series analysis.


```python
from autots import AutoTS
model = AutoTS(forecast_length=30, frequency='infer', ensemble='simple')
model = model.fit(data, date_col='Date', value_col='Close', id_col=None)
prediction = model.predict()
forecast = prediction.forecast
print(forecast)
```

Hence, using Python programming, you can utilise machine learning to forecast or predict the price of any cryptocurrency.
